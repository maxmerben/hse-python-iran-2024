### Задача 4.1. Буквы Эс

В современном персидском есть целых три буквы, которые обозначают звук /s/:

- буква «син» (**س**)
- буква «сад» (**ص**)
- буква «се» (**ث**)

Мы хотим провести исследование и изучить их статистическое распределение в текстах.

1. На вход подаётся текст. Он оформлен в виде списка, элементами которого являются абзацы текста
2. Для каждой из трёх букв необходимо подсчитать, сколько раз она встречается во всём тексте
3. Вывести статистику по каждой букве (в одну строчку)

**Внимание!** Эту задачу можно решить с помощью цикла `while`, но попробуйте использовать вместо этого только что выученный цикл `for`. Если вы уже знакомы с методом `.count()` или библиотеками `re` и `collections`, не используйте всё это при решении задач 4.1 и 4.2, попробуйте решить их с помощью инструментов, которые мы уже прошли.

Игрушечный пример со списком из трёх элементов:

```
> text = ["لباس مخصوص كار" ,"شاهنشاهی ساسانی" ,"به اثبات ثروت او رساندیم"]
«син» (س): 4, «сад» (ص): 2, «се» (ث): 2
```

____

Ниже даны переменные с каждой из трёх букв (для удобства), а также игрушечный текст из примера выше.

В самом низу страницы можно найти более крупные тексты, на которых вы сможете проверить свой код в этой и следующей задачах.

````python
sin = "س"
sad = "ص"
se  = "ث"
````

````python
text = ["لباس مخصوص كار" ,"شاهنشاهی ساسانی" ,"به اثبات ثروت او رساندیم"]
````

````python
#ваше решение
````

#### Моё решение

````python
n_sin, n_sad, n_se = 0, 0, 0

for stroka in text:
    for word in stroka.split(" "):
        for letter in word:
            if letter == sin:
                n_sin += 1
            elif letter == sad:
                n_sad += 1
            elif letter == se:
                n_se += 1

print("Текст для самопроверки 1")
print(f"«син» ({sin}): {n_sin}, «сад» ({sad}): {n_sad}, «се» ({se}): {n_se}")
````

### Задача 4.1. Буквы Эс, часть 2

В этой задаче мы также оцениваем распространённость разных букв /s/, но теперь считаем её другим способом. Здесь нам нужно посчитать не количество букв, а количество *слов*, в которых эти буквы встречаются.

К примеру, в словосочетании "شاهنشاهی ساسانی" (*šāhanšāhi-ye sāsāni*) буква «син» встречается два раза, но при этом в ней есть *всего одно слово*, в котором она встречается (*sāsāni*).

Игрушечный пример со списком из трёх элементов:

```
> text = ["لباس مخصوص كار" ,"شاهنشاهی ساسانی" ,"به اثبات ثروت او رساندیم"]
«син» (س): 3, «сад» (ص): 1, «се» (ث): 2
```

````python
# ваше решение
````

#### Моё решение

````python
n_sin, n_sad, n_se = 0, 0, 0

for stroka in text:
    for word in stroka.split(" "):
        if sin in word:
            n_sin += 1
        if sad in word:
            n_sad += 1
        if se in word:
            n_se += 1

print("Текст для самопроверки 1")
print(f"«син» ({sin}): {n_sin}, «сад» ({sad}): {n_sad}, «се» ({se}): {n_se}")
````
