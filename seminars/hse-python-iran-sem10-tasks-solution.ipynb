3. Напишите функцию **`most_frequent_words()`**. Она будет:
   - Принимать на вход два аргумента:
       - строку `text`, в которой содержится текст
       - число `min_number` — сколько раз должно встретиться слово, чтобы оно попало в частотный словарь
   - Для каждого слова в тексте подсчитывать, сколько раз оно встречается в тексте (при этом нужно очистить слова от знаков препинания по краям, а ещё нужно не учитывать регистр — например, воспринимать слова «*Тогда*» и «*тогда*» как одно слово)
   - Возвращать словарь, в котором ключами являются слова, а значениями — количество вхождений этих слов в тексте. Словарь должен быть отсортирован по алфавиту

Вот как эта функция должна работать (на примере фрагмента текста песни «Начар малай» группы Аигел):

```
> text = "Начар малай,\nНачар малай,\nЮ кулыңны,\nЮ кулыңны,\nНачар малай,\nКулыңда кем каны?\nКем каны?"

> print(most_frequent_words(text, min_number=3))
{'малай': 3, 'начар': 3}
```

````python
# ваш код для пункта 3

def most_frequent_words(text, min_number):
    text = text.split()                              # делим текст на токены (по пробелам, \t и \n)
    
    clean_tokens = []                                # пустой список для «очищенных» токенов
    for word in text:
        word = word.strip(".,:!?…-—()«»")            # для каждого токена удаляем прилипшие знаки пунктуации
        if len(word)>0:                              # если токен — не пустая строка,
            clean_tokens.append(word.lower())        # переводим его в строчный регистр и добавляем в список

    freq_dict = {}                                   # пустой словарь для пар «токен» — «кол-во вхождений»
    for word in clean_tokens:
        if word not in freq_dict:                    # если это слово ещё не встречалось в тексте,
            freq_dict[word] = 0                      # добавляем его в словарь с исходным значением (ноль)
        freq_dict[word] += 1                         # слово уже точно есть в словаре, так что прибавляем 1

    most_freq_dict = {}                              # финальный словарь (с фильтрацией и сортировкой)
    for word in sorted(freq_dict):                   # перебираем все ключи словаря, предварительно отсортировав
        if freq_dict[word] >= min_number:            # если кол-во вхождений токена больше или равно порога
            most_freq_dict[word] = freq_dict[word]   # добавляем в финальный словарь
    
    return most_freq_dict
````

**Внимание!** Протестируйте свою функцию на текстах, внимательно изучите результат и убедитесь, что она корректно очищает слова от всех ненужных знаков препинания.

6. **Сохраните словарь самых частотных слов** любого из двух текстов (только одного!) в отдельный *csv*-файл.
   - На каждой строчке такого файла должно быть записано сначала само слово, а затем через запятую (без пробелов) количество его вхождений в тексте.
   - Вы можете попробовать воспользоваться [модулем `csv`](https://pythonworld.ru/moduli/modul-csv.html) для решения этой задачи, но это необязательно.

Вот как должен выглядеть такой *csv*-файл после записи:

```
малай,3
начар,3
```

````python
# ваш код для пункта 6

with open("hpmor_freq.csv", "w", encoding="utf-8") as f:       # открываем файл в режиме записи
    
    hpmor_freq_file = []                                       # пустой список, где будут строки «токен,кол-во»
    for word in hpmor_freq:                                    # для каждого токена в словаре
        hpmor_freq_file.append(f"{word},{hpmor_freq[word]}")   # делаем строку «токен,кол-во»
    hpmor_freq_file = "\n".join(hpmor_freq_file)               # объединяем все строки символом красной строки \n
    
    print(hpmor_freq_file)
    f.write(hpmor_freq_file)
````
