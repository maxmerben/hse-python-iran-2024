## Условия

Оператор `if` позволяет нам задавать условия, при выполнении которых часть кода будет выполняться. У этого оператора такой синтаксис:

```
if a > 0:
    print("Это правда, переменная a больше нуля.")
    print("А больше и сказать-то тут нечего.")
print("Программа продолжается.")
```

После оператора `if` находится некоторое логическое условие типа `bool` (например, `a > 0` проверяет, верно ли, что переменная `a` больше нуля). Если это условие верно, то программа перейдёт к выполнению действий, написанных на строчках с отступом (например, выведет на экран какой-то текст). Если условие неверно, то программа сразу перейдёт к тому, что написано ниже строчек с отступом.

**Внимание**! После условия обязательно должно стоять двоеточие, а строчки кода, которые будут выполняться при выполнении условия, обязательно должны быть написаны с отступом. Отступ должен состоять из некоторого числа пробелов. Точное число пробелов не важно, главное — чтобы оно было одинаковым на протяжении всей программы. Однако обычно программисты ставят **отступ из четырёх пробелов**, и вам рекомендуется делать так же. Такой отступ можно легко напечатать с помощью клавиши `Tab`.

В примере ниже программа смотрит на каждое условие отдельно, и если оно выполняется, выводит на экран соответствующие строчки. В конце выводится имя — это происходит независимо от всех заданных выше условий, так как вывод имени записан без отступов:

````python
name = "Элоиза"
year = 1988

print(f"Год рождения: {year}.")

if year < 2000:
    print("Старше 2000 года рождения.")
if year < 1990:
    print("Старше 1990 года рождения.")
if year < 1980:
    print("Старше 1980 года рождения.")
print(f"Имя: {name}.")
````

Ещё несколько операторов позволяют более гибко работать с условиями. Оператор `else` даёт возможность добавить ещё один блок кода, который будет, наоборот, выполняться только в том случае, если условие в `if` не выполняется. Буквально, программа говорит: «ЕСЛИ (условие), ТОГДА (действия), А_ИНАЧЕ (альтернативные действия)».

````python
year = 1988

print(f"Год рождения: {year}.")

if year < 2000:
    print("Старше 2000 года рождения.")
else:
    print("Человек 2000 года рождения или младше.")
````

````python
year = 2003

print(f"Год рождения: {year}.")

if year < 2000:
    print("Старше 2000 года рождения.")
else:
    print("Человек 2000 года рождения или младше.")
````

То же самое в теории можно записать вот так (только с `else` проще):

````python
print(f"Год рождения: {year}.")

if year < 2000:
    print("Старше 2000 года рождения.")
if year >= 2000:
    print("Человек 2000 года рождения или младше.")
````

Оператор `elif` позволяет ввести дополнительные альтернативные условия. Их может быть одно или несколько:

````python
year = 2050

print(f"Год рождения: {year}.")

if year > 2024:
    print("Такой человек ещё не родился.")
elif year > 2000:
    print("Этот человек — молодой.")
elif year < 1935:
    print("Этот человек — пожилой.")
````

В чём отличие от простого перечисления разных условий с `if`? Название `elif` — это буквально сокращение от `else` и `if`. Это значит, что условие в `elif` будет проверяться, только если условие `if` не выполнилось. А вот если мы запишем несколько условий с помощью обычного `if`, то несколько условий могут выполниться одно за другим. Перепишем ячейку выше с помощью простых `if`’ов, и увидим, что теперь вывелось больше одной надписи. Так вышло, потому что заданный год удовлетворяет и первому (`2050 > 2024`), и второму (`2050 > 2000`) условиям:

````python
year = 2050

print(f"Год рождения: {year}.")

if year > 2024:
    print("Такой человек ещё не родился.")
if year > 2000:
    print("Этот человек — молодой.")
if year < 1935:
    print("Этот человек — пожилой.")
````

Все три оператора можно комбинировать. Они могут образовывать цепочки, в которых может быть только один `if` (в начале, обязательно) и один `else` (в конце, необязательно), а также сколько угодно `elif` (в середине, необязательно):

````python
god_postupleniya = 2023

print(f"Год поступления в Вышку: {god_postupleniya}.")

if god_postupleniya > 2024:
    print("Этот иранёнок ещё не поступил.")
elif god_postupleniya == 2019:
    print("Этот иранёнок только что выпустился.")
elif god_postupleniya < 2019:
    print("Этот иранёнок выпустился уже давно.")
else:
    print(f"Этот иранёнок на {2025-god_postupleniya} курсе.")
````

Условия можно усложнить, вложив одно в другое:

````python
food = "печенье"
drink = "молоко"

print(f"Нам предлагают еду «{food}» с напитком «{drink}».")

if food == "печенье":
    print("Вкусно. Надо есть.")
elif food == "кольраби":
    print("Непонятно что это такое. Не хочется.")
elif food == "апельсины":
    print("Аллергия. Не надо это есть.")
    
elif food == "огурцы":
    if drink == "молоко":
        print("Аллергии нет, но лучше не пробовать.")
    else:
        print("Надо брать! хрум. хрум хрум")
````

Для условий также есть специальные логические операторы `and` («и»), `or` («или»), `not` («не»).

Оператор `and` позволяет соединить несколько условий в одно большое. В таком случае действия будут выполняться, только если все эти условия будут выполняться:

````python
name = "АРДАШИР"

print("Имя:", name)

if name[0] == "А" and name[-1] == "Н":
    print("Первая буква — А, а последняя — Н. Это правда.")
else:
    print("Первая буква — А, а последняя — Н? Вовсе нет!")
````

````python
name = "АРТАБАН"

print("Имя:", name)

if name[0] == "А" and name[-1] == "Н":
    print("Первая буква — А, а последняя — Н. Это правда.")
else:
    print("Первая буква — А, а последняя — Н? Вовсе нет!")
````

Если склеить условия оператором `or`, то действия ниже будут выполняться, если хотя бы одно из условий выполняется:

````python
name = "АРДАШИР"

print("Имя:", name)

if name[0] == "А" or name[-1] == "Н":
    print("Может, первая буква — А, а может, последняя — Н. Или и то, и то.")
else:
    print("Ни первая буква — не А, ни последняя — не Н!")
````

````python
name = "КСЕРКС"

print("Имя:", name)

if name[0] == "А" or name[-1] == "Н":
    print("Может, первая буква — А, а может, последняя — Н. Или и то, и то.")
else:
    print("Ни первая буква — не А, ни последняя — не Н!")
````

С помощью `not` условие можно отрицать. Например, если мы хотим выполнить какое-то действие только тогда, когда некоторое условие неверно:

````python
name = "КСЕРКС"

print("Имя:", name)

if not name[0] == "А":
    print("Ты думал, что первая буква — А? Нет, друг, ты неправ.")
````

````python
name = "АРТАКСЕРКС"

print("Имя:", name)

if not name[0] == "А":
    print("Ты думал, что первая буква — А? Нет, друг, ты неправ.")
````

**Внимание**! Если мы соединяем несколько условий с помощью `and` или `or` и добавляем `not` к первому условию, то под отрицание попадёт только первое условие. Чтобы распространить `not` на всю комбинацию условий, надо взять эту комбинацию в скобки:

````python
name = "АРТАКСЕРКС"

print("Имя:", name)

if not name[0] == "А" and name[-1] == "Н":
    print("Если вы видите это сообщение, это значит, что выполняются два условия: (1) первая буква — НЕ буква «А», (2) последняя буква — буква «Н».")
````

````python
name = "АРТАКСЕРКС"

print("Имя:", name)

if not (name[0] == "А" and name[-1] == "Н"):
    print("Оба этих условия НЕ выполняются: (1) первая буква — буква «А», (2) последняя буква — буква «Н».")
    print("Иначе говоря: (1) первая буква — НЕ буква «А», а (2) последняя буква — НЕ буква «Н».")
````
