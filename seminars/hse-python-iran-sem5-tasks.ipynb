### Задача 5.2*. СтатиСтичеСкий анализ

В прошлой домашке мы провели первую часть нашего передового иранистического исследования: подсчитали, сколько разных букв «с» в разных персидских текстах. Теперь наши результаты сложены в список `data` (англ. «данные»). В этом списке хранятся данные по каждому изученному нами тексту. Данные по каждому тексту хранятся в виде ещё одного списка, в котором хранится три числа: число букв «син», «сад» и «се» соответственно. Получается вот такая вложенная структура:

```
data = [
    [4, 2, 2],      # текст №1:  4 буквы «син», 2 «сад»  и 2 «се»
    [100, 11, 3],   # текст №2: 100 букв «син», 11 «сад» и 3 «се»
    [73, 17, 1]     # текст №3: ...
    ...
]
```

К сожалению, так как исследованные тексты были разной длины, сравнивать эти числа между собой пока что трудно. Чтобы это исправить, нам придётся для каждого текста заменить **количество** вхождений каждой буквы на **долю** (процент) этой буквы от общего числа букв «син», «сад» и «се» в этом тексте (см. примеры ниже). Напишем программу, которая сделает это за нас!

1. На вход подаётся список с данными по каждому тексту (число текстов может быть любым). Данные для каждого текста также представлены в виде списка, в котором хранится количество букв «син», «сад» и «се» в этом тексте
2. Для каждого текста необходимо заменить количество букв «син», «сад» и «се» на долю каждой буквы от общего числа всех трёх букв в этом тексте. Таким образом, сумма долей для каждого текста будет $≈1$. Итоговые проценты должны быть округлены (не больше двух знаков после запятой)
3. Вывести на экран изменённый список

**Внимание!** Для решения этой задачи вам **не** нужно использовать переменные или код из задач 4.1 и 4.2. Эти задачи связаны только общей темой :)
