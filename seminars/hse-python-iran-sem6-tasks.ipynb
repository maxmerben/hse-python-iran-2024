### Задача 6.1. Проверка алфавита

Персидский язык использует особый вариант арабского алфавита, но в жизни для записи персидских текстов нам часто проще использовать латинский алфавит — с его помощью можно сделать тексты фонетически понятными тем, кто не читает (или плохо читает) арабскую графику. Однако единой общепринятой системы латинизации персидского алфавита не существует, и разные люди пользуются [разными системами транслитерации](https://en.wikipedia.org/wiki/Romanization_of_Persian), что часто создаёт путаницу.
